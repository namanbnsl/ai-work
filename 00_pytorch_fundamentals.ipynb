### Manipulating tensors (tensor operations)

```python
# Addition

tensor = torch.tensor([1,2,3])
tensor + 10
```

```python
# Multiply

tensor * 10
```

```python
# Subtraction

(tensor * 10) - 10
```

```python
# Division

(tensor * 100) / 10
```

```python
# Matrix multiplication (dot product)

torch.matmul(tensor, tensor), 1*1 + 2*2 + 3*3
```

```python
%%time

value = 0

for i in range(len(tensor)):
  value += tensor[i] * tensor[i]

print(value)
```

```python
# Pytorch functions are usually faster

%%time

torch.matmul(tensor, tensor)
```

```python
# shape error

tensor_1 = torch.tensor([[1,2], [3,4], [5,6]])
tensor_2 = torch.tensor([[7,10], [8,11], [9,12]])

# torch.mm(tensor_1, tensor_2) # Also torch.mm (alias for matmul)
```

```python
# To fix above errors, we can change shape using transpose

tensor_2.T
```

```python
print(tensor_2)
```

```python
# Shape error fixed

torch.mm(tensor_1, tensor_2.T)
```

### Tensor aggregation (min, max, sum, mean)

```python
# Create a tensor

x = torch.arange(0, 100, step=10)

x
```

```python
# Find the min
torch.min(x), x.min()
```

```python
# Find the max
torch.max(x), x.max()
```

```python
# Find the average/arithmetic mean - note: tensor type - float32

torch.mean(x.type(torch.float32)), x.type(torch.float32).mean()
```

```python
# Find the sum

torch.sum(x), x.sum()
```

```python
# Finding the positional min

x.argmin(), x[0]
```

```python
# Finding the positional max

x.argmax(), x[9]
```

### Reshaping, stacking, squeezing & unsqueezing tensors

```python
# Create a tensor
x = torch.arange(1, 11)

x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(2, 5)
x_reshaped, x_reshaped.shape
```

```python
# Change the view - (views share same memory as original tensor)
print(x, x.shape)
z = x.view(1, 10)

z, z.shape
```

```python
# z changes x

z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other

x_stack = torch.stack([x, x, x, x], dim=0)
x_stack
```

```python
# Squeezing

x = x_reshaped.reshape(1, 10)
x
```

```python
x_squeezed = x.squeeze()
```

```python
# Unsqueezing

print(f"Previous target: {x_squeezed}")

x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"New tensor: {x_unsqueezed}")
```

```python
# Permute - rearranges dimensions of tensor in a specified order (used for images)
# Returns view

previous_tensor = torch.tensor([1, 2, 3, 4, 5])
previous_tensor = previous_tensor.unsqueeze(dim=0)
previous_tensor.shape

previous_tensor.permute(1, 0)
```

```python
# Indexing

x = torch.arange(1, 10).reshape(1, 3, 3)
x
```

```python
x[0]
```

```python
x[0][2][0]
```

```python
# You can also use ":" to select all of a target dimension


x[0][2][2]

# Return 3, 6, 9
x[:, :, 2]
```

### PyTorch + NumPy

```python
# NumPy array to tensor

import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when converting tensor is type of numpy

array, tensor
```

```python
# Changing value of array doesn't change tensor

array = array + 1

array, tensor
```

```python
# Tensor to NumPy

tensor = torch.ones(7)
numpy_tensor = tensor.numpy()

tensor, numpy_tensor
```

```python
# Changing the tensor doesn't change array

tensor = tensor + 1

tensor, numpy_tensor
```

### Reproducibility (random out of random)

```python
import torch

rand_a = torch.rand(3, 4)
rand_b = torch.rand(3, 4)

print(rand_a)
print(rand_b)

print(rand_a == rand_b)
```

```python
# random but reproducible tensors

# set the manual seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
rand_c = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
rand_d = torch.rand(3, 4)

print(rand_c)
print(rand_d)

print(rand_c == rand_d)
```

```python
# Check access with PyTorch

import torch

torch.cuda.is_available()
```

```python
# Setup device agnostic code

device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# CPU
tensor = torch.tensor([1, 2, 3])
print(tensor, tensor.device)
```
